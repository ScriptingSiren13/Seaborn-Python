```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset('fmri')
fmri.head(3)
```

<h1 style="color:#DE006F">CREATING THE LINEPLOT</h1>

```python
sns.lineplot(x='timepoint', y='signal', data=fmri)
plt.show()
```

<h1 style="color:#DE006F">CHANGING THE AESTHETICS
</h1>

```python
sns.lineplot(x='timepoint',y='signal', hue='event', data=fmri,style='event',markers=True)
```

```python
sns.lineplot(x='timepoint',y='signal', hue='event', data=fmri,style='event',markers=True, palette=['#DE006F', 'purple'])
```
